```python
# Create a tuple of five city names
cities = ("Vizag", "hyderabad", "Varangal", "guntur", "vizayavada")
city1, city2, city3, city4, city5 = cities
print(city1)
print(city2)
print(city3)
print(city4)
print(city5)
try:
    cities[0] = "Tirupathi"
except TypeError as e:
    print("Error: Tuples are immutable, so you cannot modify elements.")
    print("Details:", e)
```

```python
python_pass = {"Vassu", "Basha", "Vyshu", "Ramsri"}
sql_pass = {"Ushasri", "Basha", "Vassu", "Keerthi"}
print("Union:", python_pass | sql_pass)
print("Intersection:", python_pass & sql_pass)
print("Python-only:", python_pass - sql_pass)
```

```python
ascii_values = [55, 56, 57, 58, 59]
for value in ascii_values:
    print(value, "->", chr(value))
    z_value = ord('Z')
print("The ASCII value of 'Z' is:", z_value)
```
